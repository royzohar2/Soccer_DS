```python
def get_data():
    with open('data/final_data.pkl', 'rb') as file:
        loaded_data = pickle.load(file)
    list_offense =[]
    for o in loaded_data:
        list_offense.append(Offense(o))

    return(list_offense)
```

```python
list_offense = get_data()
```

# Hierarchical clustering

### Perform the clustering and choosing n clusters by dendrogram

```python
sample_offense = random.sample(list_offense, 300)

linkage_matrix = hierarchical_clustering(sample_offense, 'average')
```

```python
cluster_assignments = choose_nclusters(linkage_matrix, 5)
```

### randomly choosing offense from each cluster

```python
class_1 = np.where(cluster_assignments == 1)[0]
random_1 = np.random.choice(class_1, size=min(16, len(class_1)), replace=False)

class_2 = np.where(cluster_assignments == 2)[0]
random_2 = np.random.choice(class_2, size=min(16, len(class_2)), replace=False)

class_3 = np.where(cluster_assignments == 3)[0]
random_3 = np.random.choice(class_3, size=min(16, len(class_3)), replace=False)

class_4 = np.where(cluster_assignments == 4)[0]
random_4 = np.random.choice(class_4, size=min(16, len(class_4)), replace=False)

class_5 = np.where(cluster_assignments == 5)[0]
random_5 = np.random.choice(class_5, size=min(16, len(class_5)), replace=False)
```

### plotting the random offense for each cluster

```python
offense_grid([sample_offense[i] for i in random_1], title_text='cluster 1')
offense_grid([sample_offense[i] for i in random_2], title_text='cluster 2')
offense_grid([sample_offense[i] for i in random_3], title_text='cluster 3')
offense_grid([sample_offense[i] for i in random_4], title_text='cluster 4')
offense_grid([sample_offense[i] for i in random_5], title_text='cluster 5')
```
